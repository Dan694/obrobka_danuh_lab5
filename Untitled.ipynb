```python
import cv2 as cv
import matplotlib.pyplot as plt
import pandas as pd
import os


face_cascade = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_frontalface_default.xml")
eye_cascade = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_eye.xml")

def process_image(data_row, input_folder, output_folder):
    file_name = data_row['file name']
    image_size = tuple(map(int, data_row['image size'].split("x")))
    glasses_color = data_row['glasses color']
    line_width = data_row['line width']

    input_path = os.path.join(input_folder, file_name)
    output_path = os.path.join(output_folder, f"processed_{file_name}")

    img = cv.imread(input_path)
    
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cropped_face = img[y:y+h, x:x+w]
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = cropped_face
        
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.3)
        
        eyes_cord = []

        for (ex, ey, ew, eh) in eyes:
            eyes_cord.append([ex, ey, ex + ew, ey + eh])
            cv.circle(roi_color, (ex + ew // 2, ey + eh // 2), ew // 2, (0, 0, 255), 2)
        left_eye_rect = eyes_cord[1]
        left_center_x = eyes_cord[1][2]
        left_center_y = (eyes_cord[1][3]+eyes_cord[1][1]) // 2

        right_eye_rect = eyes_cord[0]
        right_center_x = eyes_cord[0][0]
        right_center_y = (eyes_cord[0][3]+eyes_cord[0][1]) // 2
        

        cv.line(roi_color, (left_center_x, left_center_y), (right_center_x, right_center_y), (0, 0, 255), 2)


        left_center_x = eyes_cord[1][0]
        left_center_y = eyes_cord[1][2]
        
        right_center_x=eyes_cord[0][2]
        right_center_y = eyes_cord[0][3]
        
        
        cv.line(roi_color, (left_center_x, left_center_y), (left_center_x-20, left_center_y-20), (0, 0, 255), 2)
        cv.line(roi_color, (right_center_x, right_center_y-20), (right_center_x+20, right_center_y-40), (0, 0, 255), 2)

    resized_img = cv.resize(cropped_face, image_size)

    os.makedirs(output_folder, exist_ok=True)
    cv.imwrite(output_path, resized_img)
    
    plt.imshow(cv.cvtColor(resized_img, cv.COLOR_BGR2RGB))
    plt.axis('off')  
    plt.show()


n = ord("F") % 5 + 1

zavd = pd.read_excel("lab6.xlsx")
data = zavd[zavd['N'] == n]

process_image(data.iloc[0], "images", "result")
```
